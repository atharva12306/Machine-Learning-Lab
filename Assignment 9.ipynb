```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Specify the file path
file_path = 'clustering.csv'

# Check if the file exists
if os.path.exists(file_path):
    # Load the dataset
    data = pd.read_csv(file_path)
    
    # Show basic info
    print("File loaded successfully!\n")
    print("First 5 rows:\n", data.head())
    print("\nColumns:", list(data.columns))
    
    # --- Select relevant columns ---
    # (Change these column names if your CSV has different names)
    # Example: ['Annual Income (k$)', 'Spending Score (1-100)']
    x = data.iloc[:, [0, 1]]  # use first two numeric columns
    
    # --- Scatter Plot ---
    plt.figure(figsize=(8,6))
    plt.scatter(x.iloc[:, 0], x.iloc[:, 1], c='black')
    plt.xlabel(x.columns[0])
    plt.ylabel(x.columns[1])
    plt.title(f'{x.columns[0]} vs {x.columns[1]}')
    plt.show()

else:
    print(f"❌ File '{file_path}' not found in the current directory:", os.getcwd())
```
